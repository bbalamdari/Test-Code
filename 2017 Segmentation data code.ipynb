```r
#Read all data
data=read_excel("2017 National Accounts-2 Feb 2020.xlsx", col_names=TRUE, sheet="Sheet5")
head(data, n=2)
sapply(data, class)
```

```r
#Removed already those columns that had a lot of missings and here remove not needed like Number, Global Region and etc
data1=data[,-c(1,2,3,11,13,14,21,22,25,37)]
```

```r
#Chnage the column names to make it shorter
colnames(data1)[colnames(data1)=="Relationship w/ contact"] = "Contact.Relation"
colnames(data1)[colnames(data1)=="Loyal to competitor/ spot buyer/ neutral/ loyal to NC"] = "Loyalty.Competitor"
colnames(data1)[colnames(data1)=="Corrected Account Size (USD)"] = "Cor.Acc.Size"
colnames(data1)[colnames(data1)=="Actual NC Revenue USD"] = "Actual.Rev"
colnames(data1)[colnames(data1)=="Solution Preference"] = "Solution.Perf"
colnames(data1)[colnames(data1)=="Technology / product advantage and testing design favors NC                   (yes/ no)"] = "Tech.Design"
colnames(data1)[colnames(data1)=="Operator Type(Super Major vs. IOC vs. NOC vs JV)"] = "Operator.Type"
colnames(data1)[colnames(data1)=="Sales Door Opportunity"] = "Sale.Color"
colnames(data1)[colnames(data1)=="Alignment with NC OFC Strategy?(yes / no)"] = "Alignment"
colnames(data1)[colnames(data1)=="Expectation from chemical vendor"] = "Expectation"
colnames(data1)[colnames(data1)=="Connectivity:          Location          Accessibility           Competitors"] = "Connectivity"
colnames(data1)[colnames(data1)=="Resource Type"] = "Resource.Type"
colnames(data1)[colnames(data1)=="Lifecycle Stage"] = "Lifecycle.Stage"
colnames(data1)[colnames(data1)=="Chemical Capability"] = "Chemical.Capability"
colnames(data1)[colnames(data1)=="Procurement Policies"] = "Procurement.Policies"
colnames(data1)[colnames(data1)=="Contract Type"] = "Contract.Type"
colnames(data1)[colnames(data1)=="Market Segment"] = "Market.Segment"
colnames(data1)[colnames(data1)=="Formation Characteristic/ Type"] = "Formation"
colnames(data1)[colnames(data1)=="Single or multi vendor preference"] = "Vendor.Perf"
colnames(data1)[colnames(data1)=="Vendor Tier Preference"] = "Vendor.Tier.Pref"
colnames(data1)[colnames(data1)=="Needs Based Classification"] = "Need.Class"
colnames(data1)[colnames(data1)=="Digital Needs- IoT/Telemetry"] = "Dig.Need"
colnames(data1)[colnames(data1)=="Digital Needs- Data & Analytics Solutions"] = "DA.Need"
colnames(data1)[colnames(data1)=="RB TR"] = "RB.TR"
colnames(data1)[colnames(data1)=="Lifecycle Stage (groups)"] = "Lifecycle.Stage.group"

nrow(data1)
```

```r
#Change characters to factors
data1$Region=as.factor(data1$Region)
data1$Decision.Maker=as.factor(data1$Decision.Maker)
data1$Sales.District=as.factor(data1$Sales.District)
data1$Account=as.factor(data1$Account)

data1$Num.Contacts=as.factor(data1$Num.Contacts)
data1$Contact.Relation=as.factor(data1$Contact.Relation)
data1$Loyalty.Competitor=as.factor(data1$Loyalty.Competitor)
data1$Solution.Perf=as.factor(data1$Solution.Perf)
data1$Tech.Design=as.factor(data1$Tech.Design)

data1$Operator.Type=as.factor(data1$Operator.Type)
#data1$Sale.Color=as.factor(data1$Sale.Color)
#data1$Alignment=as.factor(data1$Alignment)
data1$Expectation=as.factor(data1$Expectation)
#data1$Resource.Type=as.factor(data1$Resource.Type)
data1$Formation=as.factor(data1$Formation)

data1$Lifecycle.Stage=as.factor(data1$Lifecycle.Stage)
data1$Chemical.Capability=as.factor(data1$Chemical.Capability)
data1$Procurement.Policies=as.factor(data1$Procurement.Policies)
data1$Connectivity=as.factor(data1$Connectivity)


data1$Contract.Type=as.factor(data1$Contract.Type)
data1$Vendor.Perf=as.factor(data1$Vendor.Perf)
data1$Vendor.Tier.Pref=as.factor(data1$Vendor.Tier.Pref)
data1$Need.Class=as.factor(data1$Need.Class)

data1$Dig.Need=as.factor(data1$Dig.Need)
data1$DA.Need=as.factor(data1$DA.Need)
#data1$RB.TR=as.factor(data1$RB.TR)

data1$Market.Segment=as.factor(data1$Market.Segment)
data1$Lifecycle.Stage.group=as.factor(data1$Lifecycle.Stage.group)

sapply(data1, class)
```

```r
par(mfrow=c(2,2))
Revenue.Rockies=as.vector(Rockies$Actual.Rev)
Gross.Rockies=as.vector(Rockies$GM)
OIncome.Rockies=as.vector(Rockies$OI)
DCost.Rockies=as.vector(Rockies$DC)

fun1 = function(X) {
    for (i in 1:length(X))
{
if (X[i] < 0) X[i]=-log10(abs(X[i])) else X[i]=log10(X[i])
}
    return(X)
}

DCv.Rockies=fun1(DCost.Rockies)
GMv.Rockies=fun1(Gross.Rockies)
Revenuev.Rockies=log10(Revenue.Rockies)
OIv.Rockies=fun1(OIncome.Rockies)

plot(Revenuev.Rockies,OIv.Rockies, ylab="log OI", xlab="log Revenue, $", main="Rockies, 2017", col="#0099FF")
abline(h=0, col="black", lwd=2)
plot(DCv.Rockies,OIv.Rockies, ylab="log OI", xlab="log DC, $", main="Rockies, 2017", col="blue")
abline(h=0, col="black", lwd=2)
plot(GMv.Rockies,DCv.Rockies, ylab="log DC", xlab="log GM, $", main="Rockies, 2017", col="red")
abline(h=0, col="black", lwd=2)
plot(Revenuev.Rockies,GMv.Rockies, ylab="log GM", xlab="log Revenue, $", main="Rockies, 2017", col="459")
abline(h=0, col="black", lwd=2)
```

```r
set.seed(123)
model.Rockies=lm(OI~.-1, data=Rockies[,-c(1,2,7,8,9,26,29,30,31)])
summary(model.Rockies)

```

```r
## Stepwise Regression
Rockies.both=lm(OI~.-1, data=Rockies[,-c(1,2,7,8,9,26,29,30,31)]) 
step(Rockies.both, scope=list(lower=formula(lm(OI~1, data=Rockies[,-c(1,2,7,8,9,26,29,30,31)]))),upper=formula(lm(OI~., data=Rockies[,-c(1,2,7,8,9,26,29,30,31)])), direction = "both") 
```

```r
Model.red.Rockies=lm(OI ~Num.Contacts + Contact.Relation + Loyalty.Competitor + 
    Solution.Perf + Expectation + Lifecycle.Stage + Contract.Type + 
    DA.Need - 1, data = Rockies[, -c(1,2,7,8,9,26,29,30,31)])
summary(Model.red.Rockies)
```

```r
set.seed(123)
Rockies.tree=rpart(OIsign ~., method="class", data=Rockies[,-c(1,2,7,8,9,10,26,29,30)])

cp= Rockies.tree$cptable[which.min(Rockies.tree$cptable[,"xerror"]),"CP"]
cp

Rockies.prune=prune(Rockies.tree, cp= 0.0408163265306122 )

printcp(Rockies.prune)
fancyRpartPlot(Rockies.prune, main=" Rockies 2017", cex=0.7, type=1)

```

```r
Rockies.Model.Logistic=glm(OIsign ~ Num.Contacts + Contact.Relation + Loyalty.Competitor + 
    Solution.Perf + Expectation + Lifecycle.Stage + Contract.Type + 
    DA.Need - 1, data=Rockies[,-c(1,2,7,8,9,10,26,29,30)], family = binomial)


summary(Rockies.Model.Logistic)

par(mfrow=c(2,2))
plot(Rockies$OIsign,(predict(Rockies.Model.Logistic, type="response")), main="OI Prediction, Rockies 2017", ylab="Prediction", xlab="OI", col="blue")
plot(Rockies$OI,round(predict(Rockies.Model.Logistic, type="response")),  main="OI Prediction, Rockies 2017", ylab="Prediction", xlab="OI", col="blue")

abline(v=c(0), col=c("red"), lty=c(2), lwd=c(3))
```
